<br>`type(print)` produces `builtin_function_or_method` because `print` itself is a method.

### 14. Describe and explain the result of this statement:
`type(int(float(’12.34’)))`
<br>This statement produces an output of `int`. It converts string `'12.34'` to float `12.34`, then, as it is converted to int, the decimal part is truncated, leaving an output of `12`, which is an integer.
